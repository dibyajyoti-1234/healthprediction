```python
import numpy as np
import pandas as pd
import re
import random
```

```python
df2 =pd.read_csv('health.csv')
def load_recommendations(file_path):
    
    recommendations = {}
    for _, row in df2.iterrows():
        problem = row['Disease'].strip().lower()
        recommendation = row['Description'].strip()
        if problem not in recommendations:
            recommendations[problem] = []
        recommendations[problem].append(recommendation)
    return recommendations

# Path to the CSV file
file_path='health.csv'
recommendations = load_recommendations(file_path)
df2
```

```python
def preprocess_input(user_input):
    user_input = user_input.lower()
    user_input = re.sub(r'[^\w\s]', '', user_input)
    return user_input
```

```python
def get_recommendation(user_input):
    user_input = preprocess_input(user_input)
    
    for key in recommendations:
        if key in user_input:
            # Ensure the entire response is returned
            return " ".join(recommendations[key])
    
    return "I'm sorry, I don't have a recommendation for that. Please try asking about common health issues or conditions such as headaches, fatigue, stress, or specific diseases like diabetes, hypertension, or asthma."
```

```python
def chat():
    print("Welcome to the Health Recommendation Chatbot! How can I assist you today?")
    while True:
        user_input = input("You: ")
        if user_input.lower() in ['exit', 'quit', 'bye']:
            print("Chatbot: Goodbye! Take care of your health.")
            break
        response = get_recommendation(user_input)
        print(f"Chatbot: {response}")
```
